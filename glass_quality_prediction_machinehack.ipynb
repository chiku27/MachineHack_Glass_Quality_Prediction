```python
#We can observe from the countplots that if the glass contains all the x_components from 1 to 5 then it will be of class 1
train_df1=pd.get_dummies(train_df,columns=['x_component_1','x_component_2','x_component_3','x_component_4','x_component_5','grade_A_Component_1','grade_A_Component_2','class'],drop_first=True)
train_df1.head(2)
```

```python
train_df1.shape
```

```python
x=train_df1.drop(columns=['class_2'],axis=1)
y=train_df1['class_2']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model1=RandomForestClassifier()
mod1=GridSearchCV(model1,param_grid={'criterion':['gini','entorpy'],'min_samples_leaf':[1,2,3,4,5],'n_estimators':np.arange(180,300,12)},cv=5)
mod1.fit(x_train,y_train)
```

```python
mod1.best_params_
```

```python
y_pred1=mod1.predict(x_test)
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test,y_pred1),annot=True)
```

```python
print('Training Score',mod1.score(x_train,y_train))
```

```python
print('Testing Score',mod1.score(x_test,y_test))
```
